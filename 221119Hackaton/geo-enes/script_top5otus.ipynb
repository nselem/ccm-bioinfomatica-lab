```python
#importamos pandas
import pandas as pd
#cargamos el archivo de conteos
df = pd.read_csv('EMP.5k.csv', index_col=[0])
#eliminamos columnas que no ocupamos
df2=df.drop(["Latitude"], axis = 1)
df2=df2.drop(["Longitude"], axis = 1)
#df2=df.drop(df.columns[1], axis = 1)
```

```python
df2.head(10)
```

```python
#para crear los diccionarios
rows = df2.to_dict('records')
```

```python
#importamos itemgetter
#para cada diccionario ordena los otus de mayor a menor y deja los 3 más abundantes y los coloca en un set
from operator import itemgetter

numeroOtus = 3
otusElegidos = set()
dummyRow = 0
for row in rows:
    dummyRow +=1
    listaRenglon = list()
    for otu in row.keys():
        if otu == "Unnamed: 0":
            continue
        listaRenglon.append( (otu, row[otu]) )
    renglonOrdenado = sorted(listaRenglon, key=itemgetter(1), reverse = True)
    for i in range(numeroOtus):
        otusElegidos.add( renglonOrdenado[i][0] )

len(otusElegidos)
```

```python
# seleccionamos el set de otus elegidos y los ponemos en un nuevo dataframe 
df3 = df2[otusElegidos].copy()
```
