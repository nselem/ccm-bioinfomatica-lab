```python
import pandas as pd
import plotly as plt
df = pd.read_csv("EMP.5k.CZ.csv")
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder

df2 = pd.DataFrame(df, columns=['Longitude', 'Latitude','Climate_Zone'])
lb_make = LabelEncoder()
df2['Climate_Zone'] = lb_make.fit_transform(df2['Climate_Zone'])
#df2 = df2.dropna('Climate_Zone')
kmeans = KMeans(n_clusters=11).fit_predict(df2)
import plotly.express as px
px.scatter(df2, x='Longitude', y="Latitude", color=kmeans, symbol="Climate_Zone")
```

```python
df3 = pd.DataFrame()
```

```python
Var=[]
for i in range(len(df["Climate_Zone"])):
   # print((df['Climate_Zone'][i] + "" + str(kmeans[i])))
     Var.append((df['Climate_Zone'][i] + "" + str(kmeans[i])))
```

```python
Var
```

```python
df3['NewTags'] = Var
df3
```

```python
df3['clusters'] = kmeans
df3["Climate_Zone"] = df2['Climate_Zone']
df3["Longitud"] = df2["Longitude"]
df3["Latitud"] = df2["Latitude"]
df3.head(20)
```

```python
px.scatter(df3, x='Longitud', y="Latitud", color='clusters', symbol='Climate_Zone') #NewTags = Clusters + clima(libR:KGC)
```

```python
df3.to_excel("New_Tags.xlsx")
```

```python
df['Label']= df['Climate_Zone'].replace('Cfa','WarmT-FullH-HotSum')
```

```python
df['Label']= df['Label'].replace('Dfc','Snow-FullH-CoolSum')
df['Label']= df['Label'].replace('Dfb','Snow-FullH-WarmSum')
df['Label']= df['Label'].replace('Csb','WarmT-SumDry-WarmSum')
df['Label']= df['Label'].replace('Cfb','WarmT-FullH-WarmSum')
df['Label']= df['Label'].replace('Dfa','Snow-FullH-HotSum')
df['Label']= df['Label'].replace('Am','Equat-Monsoonal')
df['Label']= df['Label'].replace('BSk','Arid-Steppe-ColdArid')
df['Label']= df['Label'].replace('Cwb','WarmT-Dessert-WarmSum')
df['Label']= df['Label'].replace('ET','Polar-PolarTundra')
df['Label']= df['Label'].replace('Af','Equat-FullH')
df['Label']= df['Label'].replace('BSh','Arid-Steppe-HotArid')
df['Label']= df['Label'].replace('BWk','Arid-Dessert-ColdArid')
df['Label']= df['Label'].replace('Dwb','Snow-WintDry-WarmSum')
df['Label']= df['Label'].replace('Csa','WarmT-SumDry-HotSum')
df['Label']= df['Label'].replace('Cwa','WarmT-WintDry-HotSum')
df['Label']= df['Label'].replace('BWh','Arid-Dessert-HotArid')
df['Label']= df['Label'].replace('Aw','Equat-WintDry')
df['Label']= df['Label'].replace('EF','Polar-PolarFrost')
df['Label']= df['Label'].replace('Dwc','Snow-WintDry-CoolSum')
```

```python
df['Label'].unique()
```

```python
Var2=[]
for i in range(len(df['Label'])):
   # print((df['Label'][i] + "" + str(kmeans[i])))
     Var2.append((df['Label'][i] + "" + str(kmeans[i])))
```

```python
df4 = pd.DataFrame()
df4['NewLabels'] = Var2
df4['clusters'] = kmeans
df4["Clima"] = df['Label']
df4["Longitud"] = df2["Longitude"]
df4["Latitud"] = df2["Latitude"]
df4.head(20)
```

```python
px.scatter(df4, x='Longitud', y="Latitud", color='NewLabels', symbol='clusters') #NewLabels = Clusters + Clima transcrito
```
